## import libraries

```python
import spacy
from spacy import displacy
```

```python
text = "Le médecin prescrit des antibiotiques au patient fiévreux après le diagnostic."
text2 = "The doctor prescribes antibiotics to the feverish patient after the diagnosis."

nlp = spacy.load("fr_core_news_sm")
nlp2 = spacy.load("en_core_web_sm")

doc = nlp(text)
doc2 = nlp2(text2)

print("\n========== FR =============\n")
for token in doc:
    print(f" {token.text} (POS) : {token.dep_}\n ")

print("\n========== ENG =============\n")
for token in doc2:
    print(f" {token.text} (POS) : {token.dep_}\n ")
```

## Search principal components

```python
print("\n========== FR =============\n")
verb_principal = [v for v in doc if v.dep_ == "ROOT"]
sujet = [t for t in verb_principal[0].lefts if t.dep_ == "nsubj"]
obj = [o for o in verb_principal[0].rights if o.dep_ == "obl:arg"]
complements = [c for c in doc if c.dep_ in ["obl:mod","nmod"]] 
print(f"verbe principal : {verb_principal[0].lemma_:<15} Sujet : {sujet[0].text:<15} Objet direct : {obj[0].text:<15}")
for complement in complements:
    print(f"Complément circonstanciel : {complement}")

print("\n========== ENG =============\n")
verb_principal2 = [v for v in doc2 if v.dep_ == "ROOT"]
sujet2 = [t for t in verb_principal2[0].lefts if t.dep_ == "nsubj"]
obj2 = [o for o in verb_principal2[0].rights if o.dep_ == "dobj"]
complements2 = [c for c in doc2 if c.dep_ == "pobj"] 
print(f"verbe principal : {verb_principal2[0].lemma_:<15} Sujet : {sujet2[0].text:<15} Objet direct : {obj2[0].text:<15}")
for complement in complements2:
    print(f"Complément circonstanciel : {complement}")
```

## Relations Extraction

```python
print("\n========== FR =============\n")
if sujet and verb_principal and obj:
    print(f"{sujet[0].text} {verb_principal[0].text} {obj[0].text}")

print("\n========== ENG =============\n")
if sujet2 and verb_principal2 and obj2:
    print(f"{sujet2[0].text} {verb_principal2[0].text} {obj2[0].text}")
```
